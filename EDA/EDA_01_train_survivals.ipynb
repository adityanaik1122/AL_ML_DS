```python
import pandas as pd
from sklearn.impute import SimpleImputer as SI

data = pd.read_csv(url)

imputer = SI(strategy='mean')
data.loc[:, ['Age']] = imputer.fit_transform(data[['Age']])

data
```

```python
data = data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
data
```

```python
imputer = SI(strategy='most_frequent')
data.loc[:, ['Fare','Embarked']] = imputer.fit_transform(data[['Fare','Embarked']])

data
```

```python
value_count = data['Sex'].value_counts()
value_count
```
