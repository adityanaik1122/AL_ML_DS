**2. Distribution Plot**

* used for univariate analysis
* used to find out for distribution
* range of teh observation
* central Tendancy
* Are there outlier?


hisplot -> Histogram
kdeplot -> Kernel density estimation
rugplot -> Marginal Distribution

```python
sns.histplot(data=tips, x="total_bill")
```

```python
sns.displot(data=tips, x='total_bill',kind='hist', bins=3)
# Bins are number of intervals
```

```python
sns.displot(data=tips, x='total_bill',kind='hist', bins=3, hue='sex')
```

```python
sns.displot(data=tips, x='total_bill',kind='hist', bins=3, hue='sex',element='step')
```

```python
sns.displot(data=tips, x='tip', col='sex',kind='hist', bins=3, hue='sex')
```

```python
#KDEPlot
# rather than using discreate bins, a KDE plot smooths the observations with gaussian kernel,producing a continuos density estimate
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill',kind='kde')
```

```python
sns.displot(data=tips, x='total_bill',kind='kde', hue='sex',fill='true')
```

```python
sns.displot(data=tips, x='total_bill',kind='kde', hue='sex',fill='true',height=10)
```

```python
sns.displot(data=tips, x='total_bill',kind='kde', hue='sex',fill='true',height=10, aspect=2)
```

**Rugplot**

* plot marginal distribution by drawing ticks along the x and y axex.
* This function is intended to complement other plots by showing the location of individual observatons in an unobtrutive way


```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

**Bivariate histogram**

* A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

**Bivariate kdeplot**

* A bivariate KDE plot smoots the (x,y) observations with a 2D guassian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

**3.Matrix Plot**


*   Haetmap
*   Clustermap

```python
# Heatmap

# plot rectangular data as color-encoded matrix
temp_df = gap.pivot(index='country', columns='year', values='lifeExp')

#axex level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# plot rectangular data as color-encoded matrix
temp_df = gap[gap['continent']=='Asia'].pivot(index='country', columns='year', values='lifeExp')

#axex level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True, linewidth=0.5,cmap='crest' )
```

```python
sns.clustermap(temp_df.iloc[:,[0,1,2,3]])
```
